# LABEL ENCODER

```python
from sklearn import tree
from sklearn.preprocessing import LabelEncoder
b = LabelEncoder()
b
```

```python
Date1 = b.fit_transform(a['Date'])
a.Date=Date1
a
```

```python
a.corr()
```

```python
b = a.drop("Date", axis =1)
b
```

# TRAIN TEST SPLIT

```python
from sklearn.model_selection import train_test_split
```

```python
x = b.drop("GLD", axis=1)
y=b["GLD"]
y
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2,random_state=42)
xtrain,xtest,ytrain,ytest
```

#  LINEAR REGRESSION  

```python
from sklearn.linear_model import LinearRegression
```

```python
e = LinearRegression()
e
```

```python
e.fit(xtrain,ytrain)
```

```python
e.score(xtest,ytest)
```

```python
e.predict(xtest)
```

#  SVM

```python
from sklearn.svm import SVR
```

```python
f = SVR()
f.fit(xtrain,ytrain)
```

```python
f.score(xtest,ytest)
```

```python
f.predict(xtest)
```

LINEAR

```python
g = SVR(kernel = 'linear')
g.fit(xtrain,ytrain)
```

```python
g.score(xtest,ytest)
```

```python
g.predict(xtest)
```

POLY

```python
h = SVR(kernel = 'poly')
h.fit(xtrain,ytrain)
```

```python
h.score(xtest,ytest)
```

```python
h.predict(xtest)
```

RBF

```python
i = SVR(kernel = 'rbf')
i.fit(xtrain,ytrain)
```

```python
i.score(xtest,ytest)
```

```python
i.predict(xtest)
```

#  DECISION TREE REGRESSOR

```python
j = tree.DecisionTreeRegressor()
j.fit(xtrain,ytrain)
```

```python
j.score(xtest,ytest)
```

```python
j.predict(xtest)
```

#  RandomForestRegressor

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
k = RandomForestRegressor(n_estimators = 60)
k.fit(xtrain,ytrain)
```

```python
k.score(xtest,ytest)
```

```python
k.predict(xtest)
```

#  XG BOOST

```python
from xgboost import XGBRegressor
n= XGBRegressor()
n.fit(xtrain,ytrain)
```

```python
n.score(xtest,ytest)
```

```python
n.predict(xtest)
```

#  VISUALISATION

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,8))
sns.heatmap(a.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.histplot(a['GLD'], kde=True, color='green')
plt.title("Distribution of GLD Prices")
plt.xlabel("GLD Price")
plt.ylabel("Frequency")
plt.show()
```

```python
top_corr = a.corr()['GLD'].abs().sort_values(ascending=False)[1:4].index.tolist()  
sns.pairplot(a[[*top_corr, 'GLD']])
plt.show()
```
